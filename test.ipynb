```python
corpus ="""Hello Welcome, to kavya's leisure study time.
Please continue learning daily! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## paragraphs--> sentence
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
sent_tokenize(corpus)
```

```python
documents=sent_tokenize(corpus)
```

```python
print(documents)
```

```python
## Tokenization
## paragraphs-->words
## sentence-->words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus) # tokenises based on punctuation e.g in kavya's the response would be ["kavya","'","s"]
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus) # tokenises such that the . in middle of the corpus is not considered as separate token. However the . at the end of paragraph is considered separate token
```
