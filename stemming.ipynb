### PorterStemmer

```python
from nltk.stem import PorterStemmer
words = ["eating","eats","eaten","writing","writes","programming","program","coding","programs","history"]
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word,"----->",stemming.stem(word))
```

```python
stemming.stem("congratulations")
```

```python
stemming.stem("sitting")
```

### RegexpStemmer
(Removes the substring that matches the passed regex)

```python
from nltk.stem import RegexpStemmer
```

```python
regex_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
regex_stemmer.stem('eating')
```

```python
regex_stemmer.stem('ingeating') # removes ing only from the end and not from front because we have given ing$
```

```python
# Remove from both prefix and suffix
regex_stemmer = RegexpStemmer('ing|s|able', min=4)
regex_stemmer.stem("ingeating")
```

```python
#Remove from pefix only
regex_stemmer = RegexpStemmer('$ing|',min=4)
regex_stemmer.stem("ingeating")
#####IT wont work
```
