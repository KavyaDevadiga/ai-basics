Lematizer is like stemming. The output of this technique is called 'Lemma' which is the root word rather than root stem

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
pos ---
Noun - n
verb - v
adjective - a
adverb - r
'''
lemmatizer.lemmatize('going',pos='n'),lemmatizer.lemmatize('going',pos='v')
```

```python
words = ["eating","eats","eaten","writing","writes","programming","program","coding","programs","history","finally","finalized","goes","fairly","sportingly"]
for word in words:
    print(word,"------>",lemmatizer.lemmatize(word,pos='v'))
```
